```
#tinyYolo
from imageai.Detection import VideoObjectDetection
import os

execution_path = os.getcwd()
tiny = VideoObjectDetection()
tiny.setModelTypeAsTinyYOLOv3()
tiny.setModelPath( os.path.join(execution_path , "yolo-tiny.h5"))
tiny.loadModel(detection_speed='flash')
plt.show()
```

```
#tinyYolo
import time
start = time.time()
video_path = tiny.detectObjectsFromVideo(input_file_path=os.path.join(execution_path, "Germany.mp4"),
                                 output_file_path=os.path.join(execution_path, "Tiny_detected"),
                                 frames_per_second=10, log_progress=True,
                                   minimum_percentage_probability=30, return_detected_frame=True,)
print(video_path)
print('sec',time.time() - start)
```
